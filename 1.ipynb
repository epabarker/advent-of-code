```python
with open('1-input.txt', 'r') as f: rotations = f.read().splitlines()
dial=50
count=0
for r in rotations:
    direction = r[0]
    magnitude = int(r[1:])
    if direction == 'R':
        passes = (dial -1 + magnitude) // 100
        dial = (dial + magnitude) % 100
    else:
        passes = ((dial - 1) // 100) - ((dial - magnitude) // 100)
        dial = (dial - magnitude) % 100
    if dial == 0: count +=1 
    count += passes
print(count)
```
