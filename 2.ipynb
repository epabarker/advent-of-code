- Find all repeating digits in range 
- Valid repetition length must be half of length of min or max
- Get length of min, length of max. 
- Get all factors of either length of min or length of max, with limit of max length / 2 (need to fit two in)
- Now we can only consider repeating patterns of length 

Now we have valid sequence lengths for each pattern. 
* If empty set, we know no valid repetitions. 
* Consider all numbers from min(lower range, smallest n digit number), where n is valid_seq_len

```python
with open('2-input.txt', 'r') as f: ranges = f.read().strip().split(',')
sum = 0

for r in ranges:
    seq_len = 0
    lower, upper = r.split('-') # Get lower and upper range 
    if len(lower) % 2 == 0: seq_len = (len(lower) // 2) # Check even length
    elif len(upper) % 2 == 0: seq_len = (len(upper) // 2) # Check even length
    print(f"Range: {r}, Sequence Length: {seq_len}")
    if seq_len == 0: continue # Skip if no valid sequence length
    
    min_n_digit = 10**(2*seq_len - 1)
    max_n_digit = (10**(2*seq_len)) - 1
    
    start = max(int(lower), min_n_digit)
    end = min(int(upper), max_n_digit)
    
    first_seq = str(start)[0:seq_len] # Get first sequence
    last_seq = str(end)[0:seq_len] # Get last sequence
    print(f"Valid Range: {start}-{end}, First Sequence: {first_seq}, Last Sequence: {last_seq}")
    for seq in range(int(first_seq), int(last_seq) + 1): # Iterate sequences
        id = int(str(seq)*2)
        if id in range(start, end + 1): # Check full number in range
            print(f"Found sequence: {str(seq)*2}") 
            sum += id
        
print(f"Total Sum: {sum}")
```
