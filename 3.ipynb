- First digit must be the max number from position 0 to k-1, where k is length of bank, as units of 10 will always be higher than units of 1
- Instead of going through the whole thing, finding max digit, then finding the max digit AFTER that digit, we should be able to iterate through one at a time 
- Set joltages to 0,0
- Consider position x. 
    - If position x is higher than max first digit, set x as max first digit and set x+1 as max second digit
    - Else if higher than max second digit, set as max second digit
- Must go from 0 - k -1

## Part 1

```python

import numpy as np

with open('3-input.txt', 'r') as f:
    banks = np.array([np.array([int(d) for d in line.strip()]) for line in f])

joltages = []
for b in banks:
    print(f"Bank: {b}")

    # Right-to-left single pass 
    best_ones = -1             # max digit to the right of current position
    best_tens = 0              # tens digit of the best pair found
    best_ones_digit = 0        # ones digit of the best pair found

    # Scan from rightmost to leftmost
    for i in range(b.shape[0] - 1, -1, -1):
        x = b[i]

        # If there exists a digit to the right, consider the pair (x, best_ones)
        if best_ones >= 0:
            # Compare: (x, best_ones) vs (best_tens, best_ones_digit)
            if (x > best_tens) or (x == best_tens and best_ones > best_ones_digit):
                best_tens = x
                best_ones_digit = best_ones

        # Update the max "ones to the right"
        if x > best_ones:
            best_ones = x

    joltage = np.array([best_tens, best_ones_digit], dtype=int)
    print(f"Joltage: {joltage[0]}{joltage[1]}")
    joltages.append(joltage)

joltages = np.array(joltages)
numbers = joltages[:, 0] * 10 + joltages[:, 1]
total_output_joltage = np.sum(numbers)
print(f"Total Output Joltage: {total_output_joltage}")
```

## Part 2

```python
def max_lex_subsequence_stack(A):
    drops = len(A) - 12
    stack = []
    for x in A:
        while stack and drops > 0 and stack[-1] < x:
            stack.pop()
            drops -= 1
        stack.append(x)
    return stack[:12]

joltages = np.array([max_lex_subsequence_stack(bank) for bank in banks])
weights = np.array([10**p for p in range(11, -1, -1)], dtype=object)
row_numbers = (joltages.astype(object) * weights).sum(axis=1)
sum(row_numbers)
```
