```python
import numpy as np
grid = np.loadtxt('4-input.txt', dtype=str)
grid = np.array([list(line) for line in grid])
grid = np.where(grid == "@", 1, 0)
```

## Part I

```python
count = 0

for x in range(grid.shape[0]):
    for y in range(grid.shape[1]):
        if grid[x,y] == 0: continue
        else:
            if grid[max(0,x-1):min(grid.shape[0],x+2),max(0,y-1):min(grid.shape[1],y+2)].sum() - grid[x,y] < 4: 
                count += 1
                
count
```

## Part II

```python
count = 0
done = False
while not done:
    current_count = 0
    for x in range(grid.shape[0]):
        for y in range(grid.shape[1]):
            if grid[x,y] == 0: continue
            else:
                if grid[max(0,x-1):min(grid.shape[0],x+2),max(0,y-1):min(grid.shape[1],y+2)].sum() - grid[x,y] < 4: 
                    count += 1
                    current_count +=1
                    grid[x,y] = 0
    if current_count == 0: done = True
    
count
```
