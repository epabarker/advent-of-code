```python
import pandas as pd
with open('5-input.txt', 'r') as f:
    ranges, ingredients = [part.split('\n') for part in f.read().split('\n\n')]
    
ranges = [pd.Interval(int(r.split('-')[0]),int(r.split('-')[1]),closed='both') for r in ranges]
ingredients = [int(i) for i in ingredients if i]
```

## Part I

```python
intervals = []
for range in sorted(ranges, key=lambda i: i.left):
    if intervals and range.left <= intervals[-1].right:
        intervals[-1] = pd.Interval(intervals[-1].left, max(intervals[-1].right, range.right), closed='both')
    else:
        intervals.append(range)
        
intervals = pd.arrays.IntervalArray(intervals)
sum([intervals.contains(i).any() for i in ingredients])
```

## Part II

```python
sum(i.length+1 for i in intervals)
```
